```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch3/members.csv")
print(df)
```

```python
# 1) 결측치수컬럼별로확인및변수에대입(시리즈)
df_cntNull = df.isnull().sum()
print(df_cntNull)
```

```python
# 2) 내림차순정렬
df_cntNull = df_cntNull.sort_values(ascending=False)
print(df_cntNull)
```

```python
# 2) 올림차순정렬
df_cntNull = df_cntNull.sort_values(ascending=True)
print(df_cntNull)
```

```python
# 3) 가장상위에있는인덱스명출력
print(df_cntNull.index[0])
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-1.csv")
print(df)
```

```python
# 1) 3사분위수와1사분위수의차이를절대값으로계산
q1 = df['age'].quantile(0.25)
q3 = df['age'].quantile(0.75)
result = abs(q1 - q3)
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch4/data4-2.csv")
print(df)
```

```python
# 1) 비율이40%보다크고50%보다작은조건
cond1 = (df['loves'] + df['wows']) / df['reactions'] > 0.4
cond2 = (df['loves'] + df['wows']) / df['reactions'] < 0.5
```

```python
# 2) type이video인조건
cond3 = df['type'] == 'video'
```

```python
# 3) 조건에맞는데이터수구하기
print(len(df[cond1 & cond2 & cond3]))
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch5/data5-2.csv")
print(df)
```

```python
# 1) bmi 파생변수계산
df['bmi'] = df['Weight'] / (df['Height']/100)**2
print(df)
```

```python
# 2) 조건1. 정상체중bmi
cond1 = (df['bmi'] >= 18.5) & (df['bmi'] < 23)
print(cond1)
```

```python
# 3) 조건2. 위험체중bmi
cond2 = (df['bmi'] >= 23) & (df['bmi'] < 25)
print(cond2)
```

```python
# 4) 조건1의인원와조건2의인원차이계산, 절대값처리
print(abs(len(df[cond1]) - len(df[cond2])))
```

```python
import pandas as pd
df = pd.read_csv("https://raw.githubusercontent.com/lovedlim/bigdata_analyst_cert/main/part4/ch7/student_assessment.csv")
print(df)
```

```python
# 1) 결측치제거
df = df.dropna()
print(df)
```

```python
id = df['id_assessment'].value_counts().idxmax()
cond = df['id_assessment'] == id
df = df[cond]
print(df)
```

```python
# 3) 과목점수스탠다드스케일
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df['score'] = scaler.fit_transform(df[['score']])
```

```python
# 4) 가장큰값
print(round(df['score'].max(), 3))
```
